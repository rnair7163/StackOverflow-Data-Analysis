```python
import pandas as pd
import numpy as np
df = pd.read_csv('post_questions_file_plotly.csv', encoding="ISO-8859-1", header = None)
```

```python
d = dict()

for tag in df[3]: 
    # Check if the tag is already in dictionary 
    if tag in d: 
        # Increment count of tag by 1 
        d[tag] = d[tag] + 1
    else: 
        # Add the tag to dictionary with count 1 
        d[tag] = 1
    
```

```python
#Create a data frame with column "Tags" and "Counts"
dfTags = pd.DataFrame(list(d.items()), columns = ["Tags", "Counts"])
```

```python
#Sort the DataFrame by column "Counts" in descending
dfTags.sort_values(by="Counts", inplace = True, ascending = False)
```

```python
dfTags.to_csv("post_questions_tagCount.csv", index=False)
```
